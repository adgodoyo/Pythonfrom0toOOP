# Diccionarios (por clave, mutable)

# {}

Son colecciones no ordenadas de pares clave-valor. Las claves son nicas dentro de un diccionario.

```python
# Definicin de un diccionario
mi_agenda = {'clave1': 1, 'clave2':"hola", 'clave3': [1,2,3]}
```
