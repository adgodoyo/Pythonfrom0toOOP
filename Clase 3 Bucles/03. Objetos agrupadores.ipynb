# Listas (ordenada - muta)

# []

Son secuencias ordenadas de elementos y son mutables. Se pueden modificar después de su creación.

Imagina que tienes un armario con muchos ganchos en una fila. Cada gancho puede tener una prenda de ropa colgada. En Python, este armario es una lista y cada prenda de ropa es un elemento. Puedes agregar más ropa, quitarla o cambiarla de lugar, y cada pieza tiene su propio número de gancho (índice) para identificarla.

```python
# Definición de una lista - Guardar una lista en una caja
mi_armario = [1, "hi", 3, 4.5, 5]
```

```python
mi_armario
```

```python
#Clase de la caja
type(mi_armario)
```

```python
type("hola")
```

```python
# Accediendo a los elementos de una lista
print(mi_armario[1]) # Imprime 1
```

```python
type(mi_armario[1])
```

```python
# Modificando elementos de una lista
nombre="Daniel"
mi_armario[0] = "Daniel"
print(mi_armario) # Imprime 10
```

```python
nombre="Daniel"
print(nombre[1])
```

# Tupla (ordenada - Inmutable)

# ()

Son similares a las listas, pero son inmutables, lo que significa que no puedes modificar su contenido una vez creadas.

Imagina un libro en el que no puedes cambiar las palabras una vez que están escritas. Aunque puedes leer las palabras (o elementos) en cualquier página (índice), no puedes simplemente borrar una palabra o agregar una nueva en medio. La tupla es ese libro impreso: una vez que has decidido qué va en él, no puedes cambiarlo.

```python
# Definición de una tupla
mi_libro = (1, 2, "asfsr", 4, 5)
```

```python
mi_libro
```

```python
type(mi_libro)
```

```python
mi_libro[2]
```

```python
mi_libro[2]=456132
```

```python
len(mi_libro) 
#La función len() en Python se utiliza para determinar el número de elementos en un objeto. 
```

# Diccionarios (por clave, mutable)

# {}

Son colecciones no ordenadas de pares clave-valor. Las claves son únicas dentro de un diccionario.

Imagina una agenda, para encontrar el número de teléfono de una persona, buscas por su nombre y encuentras el número. En un diccionario, el nombre sería la "clave" y el número de teléfono sería el "valor"(el objeto).

```python
# Definición de un diccionario
mi_agenda = {'clave1': 1, 'clave2':"hola", 'clave3': [1,2,3]}
```

```python
mi_agenda
```

```python
type(mi_agenda)
```

```python
# Accediendo a los elementos de un diccionario
mi_agenda['clave3'] # Imprime 1
```

```python
mi_agenda["clave3"]
```

```python
# Modificando elementos de un diccionario
mi_agenda['clave1'] = 10
print(mi_agenda) # Imprime 10
```
