```python
class EstudiantesDeProgramacionV2:
    def __init__(self, name, edad, eleccion_prisionero, eleccion_gallina, eleccion_sexos):
        self.__nombre = name  # Atributo privado
        self.__edad = edad  # Atributo privado
        self.__eleccion_prisionero = eleccion_prisionero # Atributo privado
        self.__eleccion_gallina = eleccion_gallina # Atributo privado
        self.__eleccion_sexos = eleccion_sexos # Atributo privado
        self.__facultad = "Facultad de Economía"  # Atributo privado
        self.__universidad = "Externado"  # Atributo privado

    def duplicar_edad(self):
        print(self.__edad*2)

    # Métodos getter para acceder a los atributos privados
    
    def get_nombre(self):
        return self.__nombre

    def get_edad(self):
        return self.__edad

    def get_facultad(self):
        return self.__facultad

    def get_universidad(self):
        return self.__universidad

    # Métodos setter para modificar los atributos privados
    def set_nombre(self, nuevo_nombre):
        self.__nombre = nuevo_nombre

    def set_edad(self, nueva_edad):
        self.__edad = nueva_edad

    # Los métodos dilema_del_prisionero, juego_de_la_gallina y batalla_de_los_sexos
    # pueden permanecer iguales o ajustarse para usar los métodos getter y setter.

```

### Resumen 

Python realiza una operación llamada "name mangling" en estos nombres. Esto significa que cuando defines un atributo o método con dos guiones bajos, Python lo renombra en el backend para incluir el nombre de la clase (_ClassName__attribute). Esto se hace para evitar conflictos de nombres en subclases y para hacer más difícil el acceso a estos atributos desde fuera de la clase.

# 3. Polimorfismo

```python
class EstudiantesDeDerecho:
    def __init__(self, name, edad):
        self.nombre = name  # Atributo privado
        self.edad = edad 

    def mi_slogan(self):
        print(f"{self.edad} años evitando las matematicas")
```

```python
class EstudiantesDeEconomia2:
    def __init__(self, name, edad):
        self.nombre = name  # Atributo privado
        self.edad = edad 

    def mi_slogan(self):
        print(f"Métale unos supuestos, todo lo demás constante y ahí la realidad te da, te lo dice {self.nombre}")
```

```python
Est1= EstudiantesDeDerecho("Juan",23)
Est2= EstudiantesDeEconomia2("Juana",24)
Est3= EstudiantesDeDerecho("Juanu",22)
Est4= EstudiantesDeEconomia2("Juano",19)
Est5= EstudiantesDeDerecho("Juani",18)
Est6= EstudiantesDeEconomia2("Juanx",17)
Est7= EstudiantesDeDerecho("Juane",16)

lista_de_Estudiantes=[Est1,Est2, Est3,Est4, Est5, Est6, Est7]
```

```python
for estudiante in lista_de_Estudiantes:
    estudiante.mi_slogan()
```

### Resumen
El polimorfismo en la Programación Orientada a Objetos (POO) es un principio que permite que objetos de diferentes clases sean tratados como objetos de una clase común. Esto se logra principalmente a través de la herencia y la interfaz común que proporcionan los métodos de la superclase. El polimorfismo permite que un mismo método se comporte de manera diferente en diferentes clases hijas, lo que significa que diferentes objetos pueden responder a la misma llamada de método de maneras distintas.


# Abstracción 

La abstracción te permite crear una estructura que representa un concepto general, sin necesidad de implementar los detalles (pero eventualmente estos deben ser precisados). 
Esto se hace usando `clases base` o interfaces que definen métodos abstractos. Un método abstracto es un método que está declarado, pero no tiene una implementación. 

Se declara para que las subclases que heredan de esta clase base esten obligadas a implementar estos métodos abstractos en el momento de su creación, proporcionando así detalles específicos que varían según la clase.

```python
from abc import ABC, abstractmethod

class Estudiante(ABC):
    def __init__(self, name, edad):
        self.__nombre = name
        self.__edad = edad

    @abstractmethod
    def actividad_extra_curricular(self):
        pass

    # Métodos getter y setter para nombre y edad
    def get_nombre(self):
        return self.__nombre

    def set_nombre(self, nuevo_nombre):
        self.__nombre = nuevo_nombre

    def get_edad(self):
        return self.__edad

    def set_edad(self, nueva_edad):
        self.__edad = nueva_edad
```

```python
class EstudianteDeMatematicas(Estudiante):
    pass
```

```python
estudiante = EstudianteDeMatematicas("Ana", 20) #Falló porque no le dimos detalle al método actividad_extra_curricular
```

```python

class EstudianteDeMatematicas(Estudiante):
    def actividad_extra_curricular(self):
        return "Participa en el club de matemáticas"
```

```python
estud= EstudianteDeMatematicas("daniel", 26)
```

```python
estud.actividad_extra_curricular()
```

### Resumen

La abstracción en Programación Orientada a Objetos (POO) es un principio que se enfoca en ocultar la complejidad detallada y mostrar solo la funcionalidad esencial al usuario. Permite al programador trabajar con conceptos de alto nivel, ocultando los detalles de bajo nivel. En POO, esto se logra mediante clases que representan abstracciones de entidades del mundo real con propiedades y comportamientos, y a través de interfaces o clases abstractas que definen 'plantillas' para otras clases.
