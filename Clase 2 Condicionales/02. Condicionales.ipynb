# if-else

En Python, la estructura condicional if y else permite ejecutar bloques de cdigo basndose en una condicin. Es una manera de decirle al programa: "Si esto es verdadero, entonces haz esto; de lo contrario, no hagas nada o haz algo diferente".
Imaginemos que alguien guard en una caja(variable) llamada x un nmero cualquiera, ej: x=5. Usted quiere programar un cdigo de manera que, si el nmero es mayor a 4, el computador seale que es mayor a 4, y seale que es igual o menor en caso contrario.

La sintaxis (forma de escribir) es de la siguiente manera:
1.	La palabra clave if.
2.	Una condicin que se evala como verdadera (True) o falsa (False).
3.	: despus de la condicin, en la misma lnea
4.	Un bloque de cdigo que se ejecuta si la condicin es verdadera (True), SIEMPRE VA CON SANGRA (tecla tab).
5.	else: (opcional)
6.	Un bloque de cdigo (opcional) que se ejecuta si la condicin es falsa (False), SIEMPRE VA CON SANGRA (tecla tab).

# if-elif-else

La declaracin **elif** en Python es una abreviatura de **"else if"**. Se usa para verificar mltiples condiciones, una tras otra, hasta que una se cumple. Si la condicin del **if** inicial no se cumple, Python pasa a la condicin **elif** siguiente y as sucesivamente. Si ninguna de las condiciones **elif** se cumple, se ejecutar el bloque de cdigo que sigue al else, si es que existe uno.
Una forma sencilla de entenderlo:

	Imagina que **if** es como preguntar, "Esto es verdad?", si s ejecuta una desterminadas acciones.

	Si no lo es, **elif pregunta**, "Qu tal esto otro, es verdad?"

	Puedes tener tantas preguntas elif como necesites, una tras otra.

	Al final, si ninguna de las preguntas anteriores fue respondida con un "s" (es decir, todas las condiciones anteriores eran falsas), **else* es como decir, "Si nada de lo anterior es verdad, entonces hagamos esto."


Es como una serie de preguntas que se hacen en orden hasta encontrar una respuesta afirmativa, y en cuanto se encuentra esa respuesta, se ejecuta el cdigo asociado y se ignoran las dems preguntas. Si no se encuentra ninguna respuesta afirmativa y hay un else, se ejecuta el cdigo en el else. Si no hay un else y ninguna condicin es verdadera, entonces no se ejecuta ninguna accin de ese bloque de cdigo.


```python
color_semaforo=input("introduce un color")

if color_semaforo == 'verde':
    accion = 'seguir adelante'
elif color_semaforo == 'amarillo':
    accion = 'prepararse para detenerse'
elif color_semaforo == 'rojo':
    accion = 'detenerse por completo'
else:
    accion = 'avisar a las autoridades que el semforo est daado'

print(accion)
```
